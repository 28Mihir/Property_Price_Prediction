```python
x = df7.drop(['price'],axis=1)
x.head()
```

Linear Regression



```python
# Linear Regression

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(x_train,y_train)

y_pred1=  lm.predict(x_test)

y_pred1
```

```python
residuals_LR = y_test-y_pred1

plt.figure(figsize=(20, 6))
plt.scatter(y_test, residuals_LR, alpha=0.5)
plt.title('Residual Plot')
plt.xlabel('Actual Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
plt.scatter(y_test,y_pred1)
```

```python
plt.hist(y_test, bins=20, alpha=0.5, label='Actual')

# Create a histogram for 'predictions' (predicted values)
plt.hist(y_pred1, bins=20, alpha=0.5, label='Predicted')

# Add labels and legend
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
mae1 = mean_absolute_error(y_test, y_pred1)
r1 = r2_score(y_test, y_pred1)

print(f'Mean Absolute Error (MAE): {mae1:.4f}')
print(f'R-squared (R2): {r1:.4f}')

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=8, test_size=0.20)
cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import RandomizedSearchCV

tree_regressor = DecisionTreeRegressor()

random_grid1 = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [2, 4, 6, 8, 10, 12, 14, 16],
    'min_samples_split': [2, 5, 10, 15, 100],
    'max_features': ['sqrt', 'log2'],
    'min_samples_leaf': [1, 2, 5, 10]
}
```

```python
random_grid1

rd_random = RandomizedSearchCV(estimator = tree_regressor, 
                               param_distributions = random_grid1,scoring='neg_mean_squared_error',
                               n_iter = 10, cv = 5, 
                               verbose=2, 
                               random_state=42, n_jobs = 5)
```

```python
rd_random.fit(x_train,y_train)
```

```python
rd_random.best_params_
```

```python
predictions_DT=rd_random.predict(x_test)
```

```python
plt.hist(y_test, bins=20, alpha=0.5, label='Actual')
```

```python
# Create a histogram for 'predictions' (predicted values)
plt.hist(predictions_DT, bins=20, alpha=0.5, label='Predicted')

# Add labels and legend
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

```python
plt.scatter(y_test,predictions_DT)
```

```python
residuals_DT = y_test - predictions_DT

plt.figure(figsize=(10, 6))
plt.scatter(y_test, residuals_DT, alpha=0.5)
plt.title('Residual Plot')
plt.xlabel('Actual Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions_DT))
print('MSE:', metrics.mean_squared_error(y_test, predictions_DT))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions_DT)))
print('R-squared', r2_score(y_test, predictions_DT))
```
